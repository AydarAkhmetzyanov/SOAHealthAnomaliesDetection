```python
import pandas as pd
df2=pd.read_csv("stat.csv",sep=';', decimal=",")
#, nrows=1000
```

```python
df2["INTERVAL_START"] = pd.to_datetime(df2["INTERVAL_START"])
```

```python
df2["ServiceContractMethod"] = df2["ServiceContract"] + " " + df2["MethodName"]
```

```python
df2.drop('DeploymentUnit', axis=1, inplace=True)
df2.drop('ServiceContract', axis=1, inplace=True)
df2.drop('MethodName', axis=1, inplace=True)
```

```python
df2.head(30)
```

```python
grouped = df2.groupby(['ServiceContractMethod','INTERVAL_START'], as_index=False)
```

```python
list(grouped.groups)[0:10]
```

```python
import numpy as np
aggregated = grouped.agg({'SucceededCount' : np.sum,
                          'BusinessErrorCount' : np.sum,
                          'TotalTime' : np.sum,
                          'AvgDuration' : np.mean})
```

```python
aggregated.head(10)
```

```python
aggregated.to_pickle('preprocessed_data.pickle')
```
